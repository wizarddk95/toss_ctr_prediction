```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_parquet('./data/train_optimized.parquet')
test = pd.read_parquet('./data/test_optimized.parquet')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 타겟 변수의 분포 확인 (비율)
click_distribution = train['clicked'].value_counts(normalize=True) * 100
print("--- Click Distribution ---")
print(click_distribution)

# 시각화
plt.figure(figsize=(8, 5))
sns.countplot(x='clicked', data=train, palette='viridis')
plt.title('Target Variable (clicked) Distribution', fontsize=15)
plt.ylabel('Count (in millions)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 이전에 최적화한 train_optimized.parquet를 사용한다고 가정합니다.
categorical_features = ['gender', 'age_group', 'inventory_id', 'day_of_week', 'hour']

# 각 카테고리 피처별 CTR 계산 및 시각화
for col in categorical_features:
    # CTR 계산 (그룹별 clicked의 평균)
    ctr_by_col = train.groupby(col)['clicked'].mean().sort_values(ascending=False)
    
    print(f"--- CTR by {col} ---")
    print(ctr_by_col) # 상위 10개만 출력
    print("\n")
    
    # 시각화 (고유값이 너무 많으면 상위 20개만)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=ctr_by_col.index, y=ctr_by_col.values, palette='viridis')
    plt.title(f'Top 20 CTR by {col}', fontsize=15)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 타겟과의 상관관계 계산
correlations = train.corr(numeric_only=True)['clicked'].abs().sort_values(ascending=False)
print("--- Top 20 Features Correlated with 'clicked' ---")
print(correlations.head(20))
```
